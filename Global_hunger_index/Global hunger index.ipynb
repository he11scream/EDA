```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
```

```python
data = pd.read_csv('Global hunger index.csv', encoding='windows-1251')
```

```python
data.head()
```

```python
data.info()
```

```python
data.columns
```

```python
data = data.drop(['Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10', 'Unnamed: 6', 'Unnamed: 11'], axis=1)
```

```python
data = data.replace({'<5':'4'})
```

```python
data = data.replace({'—': 0})
```

```python
data = data.replace({'20–34.9*':'27', '35–49.9*':'42'})
```

```python
data['2000'] = data['2000'].astype(float)
data['2007'] = data['2007'].astype(float)
data['2014'] = data['2014'].astype(float)
data['2022'] = data['2022'].astype(float)
```

```python
plt.figure(figsize=(16, 12))
sns.barplot(data=data.sort_values(by=['2000'], ascending=False).head(10), x='2000', y='Country', palette='plasma')
plt.title("2000 year", fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(16, 12))
sns.barplot(data=data.sort_values(by=['2007'], ascending=False).head(10), x='2007', y='Country', palette='plasma')
plt.title("2007 year", fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(16, 12))
sns.barplot(data=data.sort_values(by=['2014'], ascending=False).head(10), x='2014', y='Country', palette='plasma')
plt.title("2014 year", fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(16, 12))
sns.barplot(data=data.sort_values(by=['2022'], ascending=False).head(10), x='2022', y='Country', palette='plasma')
plt.title("2022 year", fontsize=20)
plt.show()
```

```python
data['dinamic'] = abs(data['2022']-data['2000'])
```

```python
plt.figure(figsize=(16, 12))
sns.barplot(data=data.sort_values(by=['dinamic'], ascending=False).head(9), x='dinamic', y='Country', palette='plasma')
plt.title("Biggest positive change", fontsize=20)
plt.show()
```

```python
data['negative'] = data['2000']-data['2022']
```

```python
plt.figure(figsize=(16, 12))
sns.barplot(data=data.sort_values(by=['negative']).head(9), x='negative', y='Country', palette='plasma')
plt.title("Biggest negative change", fontsize=20)
plt.show()
```
