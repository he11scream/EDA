```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
sns.set_style('darkgrid')
sns.set(font_scale=1.2)
```

<br>
<h3 style = "font-size: 40px; font-family:cursive; font-weight : bold; background-color: #33cc99 ; color : white; text-align: center; border-radius: 20px 20px;">EDA</h3>
<br>

```python
data = pd.read_csv('starbucks.csv')
data.head()
```

```python
data = data.drop('Unnamed: 0', axis=1)
```

```python
data.info()
```

<br>
<h3 style = "font-size: 30px; font-family:cursive; font-weight : bold; background-color: #33cc99 ; color : white; text-align: center; border-radius: 20px 20px;">Visualization</h3>
<br>

```python
type_st = data['type'].value_counts()
plt.figure(figsize=(10, 10))
plt.title('Starbacks types')
plt.pie(data=data, x=type_st.values, labels=type_st.index, autopct='%1.1f%%', explode=(0.1, 0, 0, 0, 0, 0, 0), shadow=True)
plt.show()
```

<br>
<h1 style = "color: #2F4F4F; text-align: center; font-family:cursive; font-weight : bold;">As expected, baked goods are a big part of the Starbucks menu. </h1>
<br>

```python
plt.figure(figsize=(12, 10))
sns.heatmap(data.corr())
```

```python
sns.catplot(data=data, kind='bar', y='calories', x='type', height=8).set(title="Calories in Item Type")
plt.show()
```

```python
sns.catplot(data=data, x='type', y='fat', kind='box', height=8).set(title="Fat in Item Type")
plt.show()
```

```python
sns.catplot(data=data, x='type', y='carb', kind='box', height=8).set(title='Carb in Item Type')
plt.show()
```

```python
sns.catplot(data=data, x='type', y='fiber', kind='bar', height=8).set(title="Fiber in Item Type")
plt.show()
```

```python
sns.catplot(data=data, x='type', y='protein', kind='bar', height=8).set(title='Protein in Item Type')
plt.show()
```
